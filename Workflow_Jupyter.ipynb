# Pre-Processing
From corpus to n-grams

```r
sc_1gram = load.corpus.and.parse(files = "all",
                                 corpus.dir = "corpus",
                                 markup.type= "plain",
                                 corpus.lang = "Other",
                                 splitting.rule = NULL,
                                 sampling = "no.sampling",
                                 sampling.with.replacement = FALSE,
                                 features = "w", 
                                 ngram.size = 1,
                                 preserve.case = FALSE,
                                 encoding = "UTF-8")
```

Create a list with freq., than a table

```r
sc_500MFW_1gram_l = make.frequency.list(sc_1gram, head=500)
sc_500MFW_1gram_t = make.table.of.frequencies(sc_1gram, sc_500MFW_1gram_l, relative=T)
```

```r
str(sc_500MFW_1gram_l)
```

```r
str(sc_500MFW_1gram_t)
```

Calculate Burrows' delta

```r
dist_matrix = as.matrix(dist.delta(sc_500MFW_1gram_t))
```

```r
head(dist_matrix)
```

Apply multidimensional scaling

```r
md1 = cmdscale(dist_matrix, eig = TRUE)
```

```r
str(md1)
```

```r
headers1=dimnames(md1$points)[[1]] #get the titles
```

```r
x1=unlist(md1$points[, 1]) # get the x-coord
```

```r
y1=unlist(md1$points[, 2]) # get the y-coord
```

```r
df_md1=data.frame(id=seq(49),md=1, h=headers1, x=x1, y=y1) # make a data-frame
```

```r
df_md1$category = "other"
df_md1$category[headers1=="Egla"]="Egla"
df_md1$category[headers1=="Heimskringla"] = "Hmsk"
```
